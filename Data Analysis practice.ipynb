```python
import pandas as pd
#max column width is 100 - widen the column
pd.set_option('display.max_colwidth', 100)

data = pd.read_csv("CVmessages.csv", sep='\t', header=None)
data.columns = ['message_list']

data.head()

def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

data['message_list_clean'] = data['message_list'].apply(lambda x: remove_punct(x))
data.head()
```

```python
#Tokenize
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

data['message_list_tokenized'] = data['message_list_clean'].apply(lambda x: tokenize(x.lower()))
data.head()
```

```python
#removing stop words
import nltk

stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopwords]
    return text

data['message_list_nostop'] = data['message_list_tokenized'].apply(lambda x: remove_stopwords(x))
data.head()
```

```python
#stemming
import nltk

ps = nltk.PorterStemmer()
wn = nltk.WordNetLemmatizer()

def stemming(tokenized_list):
    text = [ps.stem(word) for word in tokenized_list]
    return text
data['message_list_stemmed'] = data['message_list_nostop'].apply(lambda x: stemming(x))
data.head()
```

```python
def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

data['message_list_lemmatized'] = data['message_list_nostop'].apply(lambda x: lemmatizing(x))
data.head(10)
```

```python
#creating a single script to remove punctuation, tokenising, removing stop words, and lemmatizing

def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [wn.lemmatize(word) for word in tokens if word not in stopwords]
    return text
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(analyzer = clean_text)
x_counts = count_vect.fit_transform(data['message_list'])
print(x_counts.shape)
print(count_vect.get_feature_names())
```

```python
x_counts = pd.DataFrame(x_counts.toarray())
print(x_counts)
```

```python
x_counts.columns = count_vect.get_feature_names()
x_counts
```

```python
#vectorizing using n-grams
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = " ".join([wn.lemmatize(word) for word in tokens if word not in stopwords])
    return text

#pass it into a lambda function 
data['cleaned_messages'] = data['message_list'].apply(lambda x: clean_text(x))
data.head()
```

```python
ngram_vect = CountVectorizer(ngram_range = (2,2))
x_counts = ngram_vect.fit_transform(data['cleaned_messages'])
print(x_counts.shape)
#print(ngram_vect.get_feature_names())
```

```python
data_sample = data[0:50]
ngram_vect_sample = CountVectorizer(ngram_range = (2,4))
x_counts_sample = ngram_vect_sample.fit_transform(data_sample['cleaned_messages'])
print(x_counts_sample.shape)
print(ngram_vect_sample.get_feature_names())
```

```python
x_counts_sample_df = pd.DataFrame(x_counts_sample.toarray())
x_counts_sample_df.columns = ngram_vect_sample.get_feature_names()
x_counts_sample_df
```
